# Step One: Set up your analysis and Jupyter Notebook

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.dates as md
%matplotlib inline
from datetime import datetime
import numpy as np
import seaborn as sns
import pandas as pd
import scipy as sp
import statsmodels.api as sm
from IPython.display import Image, display
mpl.rc('figure', max_open_warning = 0)
teamdf = pd.read_csv('Teams.csv') # yearly stats and standings
cities = pd.read_csv('targetcities1.csv') # city data
payrolls = pd.read_csv('payrolls.csv') # team payrolls on opening day of that year
temps = pd.read_csv('temps.csv') # city temperature data during the baseball season
```

# Data Cleansing

```python
# drop columns we are not interested in
teamdf.drop(['lgID','divID','Ghome','AB','SF','IPouts','BPF','PPF','teamIDBR','teamIDlahman45',
           'teamIDretro','G','R','H','2B','3B','HR','BB','SO','SB','CS','HBP','RA','ER','ERA','CG','SHO',
         'SV','HA','HRA','BBA','SOA','E','DP','FP','DivWin','WCWin','LgWin','WSWin','name','park',
          '2ndteam'], inplace=True, axis=1)
```

```python
teamdf = teamdf[teamdf['yearID'] >= 1990] # filtering years we want
teamdf['yearID'] = teamdf['yearID'].astype(int) # convert columns to correct type
teamdf['Rank'] = teamdf['Rank'].astype(int)
teamdf['W'] = teamdf['W'].astype(int)
teamdf['L'] = teamdf['L'].astype(int)
teamdf['attendance'] = teamdf['attendance'].astype(int)
teamdf['in_city_since'] = teamdf['in_city_since'].astype(int)
teamdf['years_in_city'] = teamdf['years_in_city'].astype(int)

payrolls['yearID'] = payrolls['yearID'].astype(int)
payrolls['payroll'] = payrolls['payroll'].astype(int)
payrolls['payroll'] = payrolls['payroll'] / 1000000 # convert to 'millions of dollars'
payrolls['teamID'] = payrolls['teamID'].astype(str)
```

```python
mergedf =  pd.merge(teamdf, cities, on="City", how='inner') # merge city data with our main dataframe
mergedf = mergedf.merge(payrolls, how='inner', left_on=["teamID", "yearID"], right_on=["teamID","yearID"]) #team payrolls too
mergedf = pd.merge(mergedf, temps, on="City", how='inner') # bring in weather data
mergedf.drop(['State_y','State_x'], inplace = True, axis=1) # drop duplicated columns that were added durng merges
```

```python
# Wrangling some data so we can figure out how many wins teams had each year. 

mergedf['yearID'] = mergedf['yearID'].astype(int)
mergedf['yearID'] = mergedf['yearID'].astype(str) # convert yearID to string so it can concat with teamID
mergedf['yearteam'] = mergedf['yearID'] + mergedf['teamID']
mergedf['yearID'] = mergedf['yearID'].astype(int)
post = pd.read_csv('SeriesPost.csv') # W/L team data in playoffs and whether team won series or not
post['yearID'] = post['yearID'].astype(int)
post['yearID'] = post['yearID'].astype(str) # convert yearID to string so it can concat with teamID
post['yearIDteamIDwinner'] = post['yearID'] + post['teamIDwinner']
post['yearIDteamIDloser'] = post['yearID'] + post['teamIDloser']
post['yearID'] = post['yearID'].astype(int)
#new column for number of postseason wins for teams that won a post season series
serieswinnerwins = post.groupby('yearIDteamIDwinner').wins.sum()
#new column for number of postseason wins for teams that lost a post season series
seriesloserwins = post.groupby('yearIDteamIDloser').losses.sum()
#convert those to dataframes
serieswinnerwins = serieswinnerwins.to_frame()
seriesloserwins = seriesloserwins.to_frame()
#merge new dataframes with team df, many NaN for teams that did not win or make playoffs, fill with 0s.
mergedf = mergedf.merge(serieswinnerwins, how='left', left_on='yearteam', right_on='yearIDteamIDwinner')
mergedf = mergedf.merge(seriesloserwins, how='left', left_on='yearteam', right_on='yearIDteamIDloser')
mergedf[['wins','losses']] = mergedf[['wins','losses']].fillna(value=0)
#there are teams that won 1 or more series, but eventually lost a series. need to add both new columns to get total postwins
postwins = mergedf["wins"] + mergedf["losses"]
mergedf["postwins"] = postwins
```

## Creating a Linear Regression Model to Predict Attendance

### After a lot of trial and error here are the variables that created the most significant model for predicting MLB attendance:
#### rollingwins: average wins over last 3 years
#### payroll: how much the team is paying all their players in current year
#### abstemp: +/- degrees from 70F a city's average temperature is during April-September
#### parkscore: a score created from three different online rankings of best and worst ballparks (0 is the worst and 1 is the best)
#### big4teams: number of NHL, NBA, MLB, and NFL teams in the city
#### milestonextpark: distance in miles to nearest MLB team

### An R-squared value of 0.500 means that 50.0% of the variance of attendance is explained by the variables in the model. I believe if I had a larger or more informative dataset I would be able to get that value closer to 1. 

### Overall, I am pretty happy with the results. I find it interesting that my model included three variables solely tied to the city (distance to nearest team, how close the city's weather is to 70 degrees, and number of other pro sports teams). This will be important when looking at possible cities to relocate struggling MLB teams.

```python
dep_var = mergedf['attendance']
ind_var = ['rollingwins','payroll','parkscore',
          'big4teams','abstemp','milestonextpark']
mergedf = mergedf[ind_var]
mergedf['attendance'] = dep_var

x = mergedf.drop(['attendance'],axis=1)
x = sm.add_constant(x)
y = mergedf['attendance']
OLSmodel = sm.OLS(y,x)
OLSmodelResult = OLSmodel.fit()
OLSmodelResult.summary()
```

```python
# # "Each variable can be standardized by subtracting its mean from each of its values and then dividing 
# # these new values by the standard deviation of the variable. Standardizing all variables in a multiple 
# # regression yields standardized regression coefficients that show the change in the dependent variable 
# # measured in standard deviations." -wikipedia entry on standardized coefficients
from sklearn import preprocessing
# Get column names first
names = mergedf.columns
# Create the Scaler object
scaler = preprocessing.StandardScaler()
# Fit your data on the scaler object
scaled_df = scaler.fit_transform(mergedf)
scaled_df = pd.DataFrame(scaled_df, columns=names)
dep_var = scaled_df['attendance']
ind_var = ['rollingwins','payroll','parkscore',
          'big4teams','abstemp','milestonextpark']
scaled_df = scaled_df[ind_var]
scaled_df['attendance'] = dep_var

xx = scaled_df.drop(['attendance'],axis=1)
xx = sm.add_constant(xx)
yy = scaled_df['attendance']
scaled_dfOLSmodel = sm.OLS(yy,xx)
scaled_dfOLSmodelResult = scaled_dfOLSmodel.fit()
scaled_dfOLSmodelResult.summary()
```

### From the standardized regression model, we see rollingwins has the greatest impact on predicted attendance (coefficient is .41). 

```python
mpl.rcParams['figure.figsize'] = (9,5)
ax = sns.regplot(x=OLSmodelResult.predict(x), y=mergedf.attendance, data=mergedf, marker='.')
ax.set_xlabel('Model Prediction of Attendance', size=15)
ax.set_ylabel('Actual Attendance',size=15)
plt.title("Predicted Attendance vs. Actual Attendance")

plt.show()
```

### Interpretting the coefficients of the regression model
For every 1 more win a team averages over the previous 3 seasons, they can expect 31,970 more fans in the current year

For every 1 million in player payroll, a team can expect 4,064 more fans that year.

For every 1 percentile increase a stadium increases it's ranking (parkscore), the team can expect 6,397 more fans. The best park would expect to bring in 639,700 more fans than the worst park.

For every pro sports team in the city, a MLB team can expect 43,160 more fans that year.

For every 1 degree a city's average temperature during baseball season strays from 70 F, a team can expect to have 21,710 less fans that year.

For every 1 mile away from the next closest MLB team, a team will expect 495 more fans that year.
